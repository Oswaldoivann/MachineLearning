```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


df = pd.read_csv(r'C:\Users\Oswaldo Ivann\Documents\datos_inlearnin\base_datos_2008.csv', nrows = 100000)
newdf = df[['AirTime', 'DepDelay']].dropna()
```

### Como se entrena un modelo de Kmeans

```python
#KMeans(Identificamos el numero de Clusters, n_jobs=-1 mejora el rendimiento)
# .fit(DataFrame), donde se identifican nuestro archivo 
kmeans = KMeans(n_clusters=4, random_state=0, n_jobs=-1).fit(newdf)
#Mostramos a que clusters pertenecen cada uno de los datos
kmeans.labels_
```

```python
#Con esta instruccion identificamos como han quedados distribuidos nuestros datos
np.unique(kmeans.labels_,return_counts=True)
```

```python
#Mostramos nuestros datos graficamente como se distribuyen nuestros puntos
import matplotlib.pyplot as plt
plt.scatter(newdf['AirTime'], newdf['DepDelay'], c = kmeans.labels_)
plt.show()
```

```python
#Mostramos donde quedan situados los centroides de nuestros clusters identificando (x,y)
kmeans.cluster_centers_
```

```python
#Llamamos 1 millos de datos, edifinimos las columnas y eliminamos datos Nan
alldf = pd.read_csv(r'C:\Users\Oswaldo Ivann\Documents\datos_inlearnin\base_datos_2008.csv', nrows = 1e6)[['AirTime', 'DepDelay']].dropna()
```
